# Random walk
for 100 times, we throw a dice   
if it is 1 or 2, we go 1 step down  
if it is 3 or 4 or 5, we go 1 step up   
if it is 6, we throw again, and walk whatever the dice rolls up   
cannot go below step 0  
0.1% chance of falling down the stairs (start from step 0)  
can we reach step 60 ? what is the chance ?

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.rand() #random number between 0 and 1
```

```python
np.random.seed(123)
print(np.random.rand())
print(np.random.rand())
np.random.seed(123)
print(np.random.rand())
print(np.random.rand()) #number is random, but repeatable with the same seed
```

```python
all_walks = [] #initialise a list, each store a list of random_walk, i.e. this is a list of list

for i in range(500): #i times of random_walks

    random_walk = [0] #initialize list for one random walk

    for x in range(100):
        step = random_walk[-1] # set step: last element in random_walk
        dice = np.random.randint(1,7) #roll dice

        if dice <= 2:
            step = max(0, step - 1) #max(x,y) returns the bigger value, typically used for buffer certain number abv.
        elif dice > 2 and dice < 6:
            step = step + 1
        else :
            step = step + np.random.randint(1,7)
            
        if np.random.rand() <=0.001: #clumsiness
            step = 0
        
        random_walk.append(step)
        
    all_walks.append(random_walk)

#print(all_walks)

np_aw = np.array(all_walks) #convert all_walks to numpy array

np_aw_t = np.transpose(np_aw) #transpose, such that 1st row is the result of 1st dice throw of all i runs
print(np_aw_t)
```

```python
plt.plot(np_aw_t)
plt.show()
```

```python
ends = np_aw_t[-1,:] #select last row, which contains the last step of each i runs
plt.hist(ends)
plt.show()
```

```python
print(len(ends[ends>=60]))
print("The probability is "+str(len(ends[ends>=60])/500))
```
